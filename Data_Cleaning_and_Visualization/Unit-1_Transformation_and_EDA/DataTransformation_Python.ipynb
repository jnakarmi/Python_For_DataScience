## 20. Some useful summary function

a) Is there more variability in distance to some destinations than to others? Why?

```python
not_cancelled.groupby('dest').\
agg({'distance':'std'}).reset_index().sort_values('distance', ascending = False)
```

<hr style="border:1px solid black">

b) #Measures of position:Find the first and last departure for each day

```python
not_cancelled_b = not_cancelled.\
groupby(['year', 'month', 'day']).\
agg({'dep_time': ['min', 'max']}).reset_index()

not_cancelled_b.columns = ['year', 'month', 'day', 'first_dep', 'last_dep']
not_cancelled_b
```

<hr style="border:1px solid black">

c) Which destinations have the most carriers?

```python
# total number of a type of carrier in each dest.
not_cancelled_c = not_cancelled.groupby(['dest', 'carrier']).\
agg({'dep_time': 'count'}).reset_index()

not_cancelled_c.columns = ['dest','carrier','carriers']
    
not_cancelled_c.sort_values(['dest', 'carrier'], ascending = [False, True])
```

```python
# Solution
not_cancelled.groupby('dest')['carrier'].nunique()\
                      .reset_index()\
                      .rename(columns={'carrier':'carriers'}).\
sort_values(['carriers','dest'], ascending = [False,True])
```

<hr style="border:1px solid black">

d) How many flights for flown for each destination?

```python
not_cancelled.value_counts('dest')
```

<hr style="border:1px solid black">

e) What is the total number of miles a plane flew?

```python
#Total number of miles "D942DN" plane flew
not_cancelled.query('tailnum == "D942DN"')['distance'].sum()

# solution
not_cancelled.groupby('tailnum').agg({'distance': 'sum'}).reset_index()
```

<hr style="border:1px solid black">

f) How many flights left before 5am? (these usually indicate delayed flights from the previous day)

```python
not_cancelled.assign(n_early = not_cancelled.dep_time < 500).\
groupby(['year', 'month', 'day']).\
agg({'n_early' : 'sum'}).reset_index()
```

<hr style="border:1px solid black">

g) What proportion of flights are delayed by more than an hour? (Note: p = x/n)

```python
#x
not_cancelled.query('arr_delay > 60').groupby(['year','month','day']).\
agg({'arr_delay': 'count'}).reset_index().rename({'arr_delay':'x'})
```

```python
#n
not_cancelled.groupby(['year','month','day']).\
agg({'arr_delay': 'count'}).reset_index().rename({'arr_delay':'n'})
```

```python
#p
not_cancelled.assign(hour_perc = (not_cancelled.arr_delay >60)).\
              groupby(['year','month','day']).agg({'hour_perc': 'mean'}).reset_index()   
```

```python
# Using lambda function in aggregation

not_cancelled.assign(hour_perc = (not_cancelled.arr_delay >60)).\
groupby(['year','month','day']).\
agg({'hour_perc': lambda x: x.sum()/x.shape[0]}).reset_index()
```

c) Standardize to compute per group metrics. Find proportion of arrival delay of each flight per destination.

```python
popular_dests = flights.\
query('(dest == "ORD" | dest == "ATL") & (arr_delay > 0)')

popular_dests
```

```python
popular_dests.groupby(['dest'])\
.apply(lambda df: df['arr_delay']/(df['arr_delay'].sum()))\
.reset_index()\
.rename(columns = {"level_1":"index", "arr_delay":'prop_delay'})\
.merge(popular_dests.reset_index())[['year', 'month', 'day', 'dest','arr_delay','prop_delay']]
```
