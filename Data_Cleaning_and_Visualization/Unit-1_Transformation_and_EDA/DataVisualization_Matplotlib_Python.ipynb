# Matplotlib: Introduction level 

Matplotlib is the standard graphing library in python, and is typically the first graphing library a data scientist will learn when using python. It is functionally integrated with pandas and numpy for easy and efficient plotting. Furthermore, Matplotlib gives the user full control over fonts, graph styling and axes properties, though this control comes at the potential cost of lengthy blocks of code. Matplotlib is especially good for performing exploratory analysis because of the integration with pandas, allowing for quick transformations from dataframe to graph. Matplotlib is particularly good for creating basic plots like scatter plots, bargraphs and lineplots, but looks a little rough when creating more complex plots like polar scatterplots.

**Loading Needed Modules and data**

# 2. Scatterplot Arguments

MatPlotLib's Scatterplot is rigid compared to other graphical modules like seaborn. 

color mapping like `mpg %>% ggplot(mapping = aes(x = displ, y = hwy, color = class)) + geom_point()` is possible to do using Mathplotlib but the code is trickier and longer compared to ggplot. 
    

* **c or color** : set your own color for each scatter plot with the color or the c argument
* **s**: change the size of the dots with the s argument.
* **alpha**: blending value, between 0 (transparent) and 1 (opaque).

Click the link [here_1](https://www.w3schools.com/python/matplotlib_scatter.asp) or [here_2](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.scatter.html#matplotlib.axes.Axes.scatter) for all other arugments for mathplotlib's scatter function.

```python
#c or color : set your own color for each scatter plot with the color or the c argument
fig, ax = plt.subplots()

ax.scatter(x = "displ", y = 'hwy', c='red', data = mpg)

plt.show()
```

<hr style="border:1px solid black">

```python
#alpha: blending value, between 0 (transparent) and 1 (opaque).
fig, ax = plt.subplots()

ax.scatter(x = "displ", y = 'hwy', alpha=0.2, data = mpg)

plt.show()
```

<hr style="border:1px solid black">

```python
#size: change the size of the dots with the s argument.
fig, ax = plt.subplots()

ax.scatter(x = "displ", y = 'hwy', s=0.2, data = mpg)

plt.show()
```

# 3. Adding labels and graph title

```python
#theme
plt.style.use('ggplot')

fig, ax = plt.subplots()

ax.scatter(x = "displ", y = 'hwy', c='red', data = mpg)

# Customize the x-axis label
ax.set_xlabel("Displacement")

# Customize the y-axis label
ax.set_ylabel("Highway MPG")

# Add the title
ax.set_title("Relationship Between Displacement and Highway MPG", fontweight="bold")

#ax.text(2, 4, 'It looks like a negative linear relationship', style='italic',
#       bbox={'facecolor': 'red', 'alpha': 0.5, 'pad': 5})

#ax.annotate('annotate', xy=(2, 1), xytext=(3, 4),
#            arrowprops=dict(facecolor='black', shrink=0.05))
ax.text(0.99, -0.2, 'There seems be a linear relationship between them',
        horizontalalignment='right',
        transform=ax.transAxes,
        color='blue', fontsize=15)

ax.legend()

plt.show()
```

### b)  Histogram

```python
#Default matplotlib setting

import matplotlib as mpl
mpl.rcParams.update(mpl.rcParamsDefault)

fig, ax = plt.subplots()

ax.hist(x = "displ", color='pink', edgecolor='black',data = mpg) #color instead of c here

# Customize the x-axis label
ax.set_xlabel("Displacement")

plt.show()
```

<hr style="border:1px solid black">

### c)  Bar Graph with "xlabel" rotation

```python
mpg_bar = mpg['class1'].value_counts().reset_index().rename(columns = {'index': 'class1', 'class1': 'freq'})

fig, ax = plt.subplots()

ax.bar(x = "class1", height = "freq", color='green', edgecolor='black', data = mpg_bar) #color instead of c here

# Customize the x-axis label
ax.set_xlabel("Class")

# Customize the y-axis label
ax.set_ylabel("Frequency")

ax.set_xticklabels(mpg_bar["class1"], rotation = 90)

plt.show()
```

<hr style="border:1px solid black">

### d)  Boxplot

```python
fig, ax = plt.subplots()

ax.boxplot(x = "hwy", data = mpg, vert = False) #color instead of c here

# Customize the x-axis label
ax.set_xlabel("Highway MPG")

plt.show()
```

<hr style="border:1px solid black">

### e) Violin Plot

```python
fig, ax = plt.subplots()

ax.violinplot(dataset = mpg["hwy"], vert = False) #color instead of c here

# Customize the x-axis label
ax.set_xlabel("Highway MPG")

plt.show()
```

### Error bars 

Error bars are graphical representations of the variability of data and used on graphs to indicate the error or uncertainty in a reported measurement.

```python
fig, ax = plt.subplots()

# Add a bar for the fuel type "Highway MPG" column mean/std
ax.bar("P", mpg.query('fl == "p"')["hwy"].mean(),  yerr=mpg.query('fl == "p"')["hwy"].std())
ax.bar("R", mpg.query('fl == "r"')["hwy"].mean(),  yerr=mpg.query('fl == "r"')["hwy"].std())


# Label the y-axis
ax.set_ylabel("Highway MPG")

plt.show()
```

### Conditional Distribution

```python
# More complicated than in R
from matplotlib.colors import LinearSegmentedColormap

plt.style.use('ggplot')

# Conditional Distribution
df_new = diamonds[['cut', 'clarity']].pivot_table(columns="cut", 
                                                  index="clarity", 
                                                  aggfunc=len, fill_value=0).T.apply(lambda x: x/sum(x), axis=1)

# Color Map
colors = sns.color_palette("ch:start=.2,rot=-.3", n_colors=len(df_new.columns))
cmap1 = LinearSegmentedColormap.from_list("my_colormap", colors)


# plot
ax = df_new.plot.bar(stacked=True, colormap=cmap1)
ax.set_xticklabels(ax.get_xticklabels(), rotation=0)

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), title="clarity", framealpha=0)


plt.show()
```
